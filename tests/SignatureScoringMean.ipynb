```python
sig = pd.read_csv("hallmark.csv", index_col=0)
sig
```

```python
score_df = pd.DataFrame(index=df.columns)
gene_list = df.index.to_list()

for x in sig.columns:
    genes = sig[x].dropna().to_list()
    temp = set(genes).intersection(gene_list)
    if len(temp)<len(genes):
        print("Dropping genes",set(genes).difference(temp),"from gene signature ", x)
    score_df[x] = df.loc[temp,:].mean(axis=0)    
```
